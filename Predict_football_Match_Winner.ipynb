```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading the data
matches = pd.read_csv('https://raw.githubusercontent.com/dataquestio/project-walkthroughs/master/football_matches/matches.csv', index_col=0)
```

```python
#preparing the data
matches["date"] = pd.to_datetime(matches["date"])
```

```python
#converting the venue & opponent
matches['venue_code'] = matches['venue'].astype("category").cat.codes
matches['opp_code'] = matches['opponent'].astype("category").cat.codes
```

```python
#changing the weekdays to codes
matches['day_code'] = matches['date'].dt.dayofweek
```

```python
#changing the hours to integer
matches['hour'] = matches['time'].str.replace(":.+", "", regex=True)
```

```python
#creating target value
matches['target'] = (matches['result']=='W').astype(int)
```

```python
matches
```

```python
#creating machine learning model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#training data
train = matches[matches['date']<= '2022-01-01']
```

```python
#testing data
test = matches[matches['date']> '2022-01-01']
```

```python
#training the model
predictors = ['venue_code', 'opp_code', 'day_code', 'hour']

model.fit(train[predictors], train['target'])
```

```python
#predicting the data
preds = model.predict(test[predictors])
```

```python
#measuring performance with accuracy_score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test['target'], preds)
```

```python
accuracy
```

```python
#creating a dataframe with predicted result and the target values of the test dataset
combined = pd.DataFrame(dict(actual=test['target'], prediction=preds))
```

```python
#checking the crosstab
pd.crosstab(index=combined['actual'], columns=combined['prediction'])
```

```python
#measuring performance with precision score
from sklearn.metrics import precision_score
precision_score(test['target'], preds)
```

```python
#improving the precision score using rolling averages
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
#Creating a function for rolling averages
def rolling_averages (group, cols, new_cols):
  group = group.sort_values("date")
  rolling_stats = group[cols].rolling(3, closed='left').mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset=new_cols)
  return group
```

```python
#creating the cols & new_cols variables
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
#applying rolling_averages function in the groupby data for every team
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("team")
```

```python
#index doesn't go that high
matches_rolling
```

```python
#index needs fixing by making all the index values unique as a lot of values of this dataframe index are similar
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
#fixed matches rolling
matches_rolling
```

```python
#putting the machine learning model codes in a function
def make_predictions(data, predictors):
  train = data[data['date']<= '2022-01-01']
  test = data[data['date']<= '2022-01-01']
  model.fit(train[predictors], train['target'])
  preds = model.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test['target'], prediction=preds))
  precision = precision_score(test['target'], preds)
  return combined, precision
```

```python
#running the new data through the ML function
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
